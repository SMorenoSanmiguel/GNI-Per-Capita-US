###Ejercicio A
Obtener los 10 primeros países con el mayor valor per cápita del último año disponible (país, año, valor per cápita)

```
#Usando la libreria BeautifulSoup se abre el archivo
with open("/dbfs/FileStore/tables/API_NY_GNP_MKTP_PC_CD_DS46_en_xml_v2_2303962.xml") as fp:
  soup = BeautifulSoup(fp)
  
#Se crea un diccionario vacio con 4 columnas para usarse en el pandas dataframe
data_dictionary = {'Country or Area':[], 'Year':[], 'Item':[], 'Value':[]}

#Se alimenta al diccionario con los datos del archivo .xml
for record in soup.find_all('record'):
  for record in record.find_all('field'):
    data_dictionary[record['name']].append(record.text)

#Se crea un pandas dataframe desde el diccionario
df = pd.DataFrame.from_dict(data_dictionary)
df = df.pivot(index='Country or Area', columns='Year', values='Value')
df.reset_index(level=0, inplace=True)
display(df)
```

```
#Se crea una lista con los nombres de los años y se deja a un lado la columna de los nombres de los paises
cols = df.columns.drop('Country or Area')

#Se cambia el tipo de dato a numerico para que uno de los años
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
df.dtypes
```

```
#Se crea una nueva pandas dataframe que se ordena de mayor a menos en la columna del año 2019
new_df = df.sort_values(by='2019',ascending=False)
display(new_df)
```

```
#Se crea una nueva pandas dataframe que solo tiene dos columnas y 10 filas, donde se muestran los paises que en el año 2019 tiene el mayor GNI per Capita
first_10 = new_df.loc[:,['Country or Area','2019']].head(10)
display(first_10)
```

###Ejercicio B
Elegir 5 países y obtiene el delta de valor per cápita con respecto al año anterior. Años seleccionados a criterio (país, año actual, valor actual, año anterior, valor anterior, delta)

```
#Se eligen 5 paises de toda la lista
paises = ['Argentina', 'Brazil', 'Chile', 'Colombia', 'Mexico']

#Se crea una nueva pandas dataframe con solo los paises de la lista que se acaba de seleccionar
paises_df = df.loc[df['Country or Area'].isin(paises)]
display(paises_df)
```

```
#Se crea una nueva pandas dataframe que tendra solo 3 columas, que seran el nombre del pais y los años 2019 y 2018
paises_delta_df = paises_df.loc[:,['Country or Area','2018','2019']]

#Se crea el calculo estadistico que sera el completemento de la pandas dataframe recien creada
delta = (paises_delta_df['2019'] - paises_delta_df['2018'])

#Se crea una nueva columna con el valor de delta que se calculo
paises_delta_df['delta'] = delta
display(paises_delta_df)
```

###Ejercicio C
Obtiene la media, moda y mediana de valores per cápita de 5 países a elección (país, media, moda, mediana)

```
#Se crea una copia de una pandas dataframe creada anteriormente, esta nueva pandas dataframe servira para hacer el calculo de los estadisticos que se requieren
paises_agg_df = paises_df.copy()

#Se calculan los tres estadisticos solicitados
df_mean = paises_agg_df[cols].mean(axis=1)
df_mode = paises_agg_df[cols].mode(axis=1)
df_median = paises_agg_df[cols].median(axis=1)

#De aqui se puede ver que el estadistico de la media y la mediana se pueden calcular correctamente, pero tambien se observa que el estadistico de la moda no se puede obtener ya que todos los valores son diferentes
print(df_mean)
print(df_mode)
print(df_median)
```

```
#Se crean nuevas columnas para los estadisticos de la media y mediana. Se decidio no incluir columnas para la moda ya que no era posible incluirlo en una sola columna
paises_agg_df['mean'] = df_mean
paises_agg_df['median'] = df_median
#paises_agg_df['mode'] = df_mode

display(paises_agg_df)
```

```
#Se dejan solamente las columnas para los estadisticos y los paises seleccionados
paises_agg_df = paises_agg_df.loc[:,['Country or Area','mean','median']]
display(paises_agg_df)
```
